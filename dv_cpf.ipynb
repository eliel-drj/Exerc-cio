```python
cpf = str(input("Digite seu CPF: "))

um = int(cpf[0]) * 10
dois = int(cpf[1]) * 9
tres = int(cpf[2]) * 8
quatro = int(cpf[3]) * 7
cinco = int(cpf[4]) * 6
seis = int(cpf[5]) * 5
sete = int(cpf[6]) * 4
oito = int(cpf[7]) * 3
nove = int(cpf[8]) * 2

soma1 = um + dois + tres + quatro + cinco + seis + sete + oito + nove

resto1 = soma1 % 11

digito1 = 11 - resto1
print(digito1)

if digito1 < 2:
    digito1 = "0"

um = int(cpf[0]) * 11
dois = int(cpf[1]) * 10
tres = int(cpf[2]) * 9
quatro = int(cpf[3]) * 8
cinco = int(cpf[4]) * 7
seis = int(cpf[5]) * 6
sete = int(cpf[6]) * 5
oito = int(cpf[7]) * 4
nove = int(cpf[8]) * 3
dez = int(cpf[9]) * 2

soma2 = um + dois + tres + quatro + cinco + seis + sete + oito + nove + dez

resto2 = soma2 % 11

digito2 = 11 - resto2
print(digito2)

if digito2 < 2:
    digito2 = "0"

if digito1 == int(cpf[9]) and digito2 == int(cpf[10]):
    print("CPF Válido")
else:
    print("CPF Inválido")
```
